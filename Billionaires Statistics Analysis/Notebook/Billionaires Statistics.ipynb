## Requirements

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('whitegrid')
import warnings 
warnings.filterwarnings('ignore')
```

## Loading Dataset

```python
df=pd.read_csv('Billionaires Statistics Dataset.csv')
```

```python
df #Reading dataset
```

## Data Cleaning & Transformation

```python
dfc=df.copy()  #We have made a copy of our dataset and stored it in dfc dataframe
```

```python
dfc.head()
```

```python
dfc.tail()
```

```python
dfc.info()
```

#### Some of the datatype is not correct

1) 'birthDate' datatype is object
   //it is not correct
2) 'date' datatype is object
   //it is not correct
3) 'gdp_country' is object
   //it is not correct   

```python
# Changing Data Types

dfc['birthDate'] = pd.to_datetime(dfc['birthDate'])
dfc['date'] = pd.to_datetime(dfc['birthDate'])
dfc['gdp_country'] = pd.to_numeric(dfc['gdp_country'].replace('[\$,]', '', regex=True), errors='coerce')
```

```python
dfc['gdp_country']
```

```python
# Here values are coming in scientific form instead of normal form
# Lets change the values to normal form for whole dataset at a time

pd.options.display.float_format = '{:,.0f}'.format
```

```python
dfc['gdp_country']
```

```python
dfc.info()
```

```python
#Adding a new coumn 'billionaires_name' in dataset by using firstName & lastName columns

dfc['billionaires_name'] = dfc['firstName'] + ' ' + dfc['lastName']
```

```python
#Dropping similar columns from dataset

dfc.drop('personName',axis=1,inplace=True)  #personName & newly made 'billionaires_name' columns are same
dfc.drop('firstName',axis=1,inplace=True)
dfc.drop('lastName',axis=1,inplace=True)
dfc.drop('industries',axis=1,inplace=True)  #category & industries columns are same
dfc.drop('date',axis=1,inplace=True)  #date & birthDate columns are same
```

```python
dfc.columns
```

```python
dfc.shape
```

```python
dfc.describe()
```

```python
dfc.columns.nunique()
```

```python
dfc.isnull().sum()
```

#### Comment:
Lots of columns have null values, we need to address it somehow <br>
We don't delete all null values as the data gets lost <br>
We will delete the null values for the selected columns only, whose overall contribution in the data is less than 7%

```python
columns_to_replace = ['organization', 'title', 'state', 'residenceStateRegion']
dfc[columns_to_replace] = dfc[columns_to_replace].fillna('Unknown')
```

```python
dfc.dropna(inplace=True)
```

```python
dfc.isnull().sum()
```

### Geographical Analysis

#### 1) Which countries have the highest concentration of billionaires?

```python
grouped_personName_by_country=dfc.groupby('country')['billionaires_name'].count().sort_values(ascending=False)
grouped_personName_by_country.head()
```

```python
df_personName_country
```

```python
#Top 10 Countries which have highest Billionaires

plt.figure(figsize=[3,3])
sns.barplot(x="billionaires_name",y='country',data=df_personName_country.head(10)) #head(10) is used for getting Top 10 Countries
plt.title('Top 10 Countries')
plt.xlabel('Billionaires Count')
plt.ylabel('Country')
```

#### 2) Is there any correlation between a country's GDP and the no. of Billionaires?

```python
grouped_gdp_by_country=dfc.groupby('country')['gdp_country'].mean().sort_values(ascending=False).reset_index()
grouped_gdp_by_country
```

```python
plt.figure(figsize=[3,3])
sns.barplot(x="gdp_country",y='country',data=grouped_gdp_by_country.head(10)) #head(10) is used for getting Top 10 Countries
plt.title('Top 10 Countries by GDP')
plt.xlabel('Country GDP')
plt.ylabel('Country')
```

```python
#To make it more simpler, we will combine the plots 

#Billionaires Count by Country
plt.figure(figsize=[3,3])
sns.barplot(x="billionaires_name",y='country',data=df_personName_country.head(10)) #head(10) is used for getting Top 10 Countries
plt.title('Top 10 Countries by No. of Billionaires')
plt.xlabel('Billionaires Count')
plt.ylabel('Country')


#Country GDP by Country
plt.figure(figsize=[3,3])
sns.barplot(x="gdp_country",y='country',data=grouped_gdp_by_country.head(10)) #head(10) is used for getting Top 10 Countries
plt.title('Top 10 Countries by GDP')
plt.xlabel('Country GDP')
plt.ylabel('Country');
```

Here, from both the graphs we can see that 'Top 10 Countries by Billionaires' & 'Top 10 Countries by GDP' are not same in sequence. <br>
So, we can finally conclude that there is some relation between both but it is not directly correlated.
